# About the Dataset
HEALTH CARE DATASET (https://www.kaggle.com/datasets/prasad22/healthcare-dataset/data)

This is a dummy dataset  from Kaggle. This synthetic healthcare dataset has been designed to mimic real-world healthcare data and would be used for exploratory data analysis (EDA) purposes in health care analytics.

Dataset Information: The information provided on the columns highlight the columns used in this analysis.


- Age: The age of the patient at the time of admission, expressed in years.
- Gender: Indicates the gender of the patient, either "Male" or "Female."
- Blood Type: The patient's blood type, which can be one of the common blood types (e.g., "A+", "O-", etc.).
- Medical Condition: This column specifies the primary medical condition or diagnosis associated with the patient, such as "Diabetes," "Hypertension," "Asthma," and more.
- Date of Admission: The date on which the patient was admitted to the healthcare facility.
- Doctor: The name of the doctor responsible for the patient's care during their admission.
- Billing Amount: The amount of money billed for the patient's healthcare services during their admission. This is expressed as a floating-point number.
- Admission Type: Specifies the type of admission, which can be "Emergency," "Elective," or "Urgent," reflecting the circumstances of the admission.
- Discharge Date: The date on which the patient was discharged from the healthcare facility, based on the admission date and a random number of days within a realistic range.
- Test Results: Describes the results of a medical test conducted during the patient's admission. Possible values include "Normal," "Abnormal," or "Inconclusive," indicating the outcome of the test.

```python
# converting the "Date of Admission" and "Discharge Date" to date time objects

# for date of admission
health_dataframe['Date of Admission'] = pd.to_datetime(health_dataframe['Date of Admission'])

# for discharge date
health_dataframe['Discharge Date'] = pd.to_datetime(health_dataframe['Discharge Date'])
```

```python
health_dataframe.info()
```

```python
# create a new column and store the duration between date of admission and discharge date
health_dataframe['Admission Duration'] = health_dataframe['Discharge Date'] - health_dataframe['Date of Admission']
health_dataframe['Admission Duration']
```

```python
# convert Admission Duration from timedelta object to int
# a time delta 64 is a Numpy data type that stores time duration in a nanoseconds precise manner.  
# It is part of NumPy's datetime64 module, which extends the basic date and time capabilities.
# it is useful for allowing arithematic operations. calculations can be performed between intervals.
```

```python
# to ensure seamless visualisation,statistical methods and compatibility with libraries or functions on further analysis, 
# timedelta data type is converted to int. The number of days is extracted from the duration
health_dataframe['Admission Duration'] = health_dataframe['Admission Duration'].dt.days
health_dataframe['Admission Duration'].info()
```

```python
health_dataframe.head()
```

```python
# converting billing amount to int
health_dataframe['Billing Amount'] = health_dataframe['Billing Amount'].astype('int')
health_dataframe['Billing Amount'].info()
```

###### By Age

```python
# minimum and maximum ages
print(health_dataframe.Age.min())
print(health_dataframe.Age.max())
```

```python
# was billing amount affected by age?

# create a column grouping age 
def group_age(Age):
    if Age >= 80:
        return '>80'
    if Age >= 70:
        return '70-79'
    if Age >= 60:
        return '60-69'
    if Age >= 50:
        return '50-59'
    if Age >= 40:
        return '40-49'
    if Age >= 30:
        return '30-39'
    if Age >= 20:
        return '20-29'
    elif Age >= 10:
        return '<20'
```

```python
# apply function to age column in dataset
health_dataframe['Age Group'] = health_dataframe['Age'].apply(group_age)
health_dataframe['Age Group']
```

```python
health_dataframe.head()
```

```python
# goup by, sum and count billing amount
billing_age = health_dataframe.groupby(['Age Group'])['Billing Amount'].agg([np.sum, np.size]).reset_index()
billing_age
```

```python
# renaming columns
billing_age = billing_age.rename(columns = {'sum' : 'Billing Amount',
                                  'size' : 'Patients'})
billing_age
```

```python
# set the figure higher than the default
plt.figure(figsize = (10,5))

# age group order for plot
age_order = ['<20', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '>80']
# plot a bar chart 
sns.barplot(x = billing_age['Age Group'], y = billing_age['Billing Amount'], palette = 'rocket', order= age_order)
plt.title('Billing Amount by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Billing Amount')
```

Billing amount by age group has a normal distribution. 
- Age 70-79 had the highest billing amount while ages below 20 had the least.
- Patients within the lowest billing amount had the lowest number of admission.
